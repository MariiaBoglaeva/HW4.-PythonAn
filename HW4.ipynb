```python
sns.jointplot(x=house['price'], y=house['bedrooms'], kind='reg');
```

Четко выражена прямая линейная зависимость между состоянием дома и его стоимостью

```python
matrix_corr = house.corr()
matrix_corr = np.round(matrix_corr, 1)
matrix_corr
```

```python
plt.figure(figsize=(10, 8))

sns.set(font_scale=0.9)

sns.heatmap(matrix_corr, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Матрица корреляции');
```

С помощью тепловой карты матрицы корреляции можно выявить, что на показатель цены влияют состояние дома (оценка), 
жилая площадь, sqrt_above, кол-во bathrooms, вид.
